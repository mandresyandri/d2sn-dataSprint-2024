```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Charger les données
gdf_gare = gpd.read_file('datasets/last_data/gdf_gare.geojson')
gdf_region = gpd.read_file('datasets/last_data/gdf_region.geojson')
```

```python
# Définir le CRS
gdf_gare = gdf_gare.set_crs(epsg=4326)
gdf_region = gdf_region.set_crs(epsg=4326)
```

```python
# Créer la carte choroplèthe
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
_ = gdf_region.plot(column='Prix moyen par km', ax=ax, legend=True,
                legend_kwds={'label': "Tarif par km",
                             'orientation': "horizontal"})

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.set_xticks([])
ax.set_yticks([])

# Afficher la carte
plt.title("Choropleth Map of Tarif par km")
plt.show()
```
